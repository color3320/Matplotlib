### Line Graphs

```python
#here we can take year as x axis and countries gas prices as y axis
plt.plot(df.Year,df.USA)
plt.plot(df.Year,df.Canada)
#if any column has space in it's name you can -
plt.plot(df.Year,df["South Korea"])
plt.show()
```

```python
#To quickly plot a graph for all the countries use loops

plt.figure(figsize = (12,7))

plt.title("Gas prices in USD over years",fontdict = {"fontsize": 18})

for i in df.columns[1:]:
    plt.plot(df['Year'] , df[i] , label=i , marker='.')

plt.xlabel("Year",fontdict = {"fontsize": 18}) 
plt.ylabel("USD",fontdict = {"fontsize": 18})

plt.xticks(df.Year[::2])
    
plt.legend()

plt.show()
```

#### Lets make a pie chart on % of people who play with left foot and % of people who play with right foot 

```python
left = fifa.loc[fifa["Preferred Foot"]=="Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"]=="Right"].count()[0]
print(right)
print(left)
#To plot a pie chart we need numbers 
```

```python
plt.pie([left,right])
plt.show()
```

```python
labels = ["Left Foot","Right Foot"]

colors = ["#a55d45","#aab3f4"]

plt.pie([left,right],labels = labels,colors = colors)
plt.show()
```

```python
#To show value on the pie chart use autopct
plt.pie([left,right],labels = labels,colors = colors, autopct = "%.2f %%")
plt.title("Foot Preferance of Fifa Players")
#use 2 % sign to actually show %tage
plt.show()
```

```python
fifa.Weight
#as piecharts need only number we have to remove the lbs notations 
#and convert it to int or float
```

```python
fifa.Weight = [x.strip("lbs") if type(x)==str else x for x in fifa.Weight]
#here if statement because if it is not string or empty value we want to leave
#it as it is
fifa.Weight
```

```python
fifa.Weight[0] #it is still a string value
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
```

#### Now lets divide data in 5 parts

```python
light = fifa.loc[fifa.Weight<125].count()[0]
light_med = fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
med_heavy = fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]

print (light)
print(light_med)
print(medium)
print(med_heavy)
print(heavy)

weights = [light,light_med,medium,med_heavy,heavy]
```

```python
labels = ["Under 125","Between 125 & 150","Between 150 & 175","Between 175 & 200","Above 200"]

plt.pie(weights,labels = labels)
plt.show()
```

```python
#To change colors from default we can use the color scheme of ggplot
plt.figure(figsize = (12,7))

plt.title("Weight distribution of FIFA players")

plt.style.use("ggplot")

explode=(.4,.2,0,0,.4)
#explode sperates pie chart

plt.pie(weights,labels = labels,autopct = "%.2f %%",explode=explode)
plt.show()
```

### Box Chart
#### used to compare things statistically

```python
fifa.head()
```

```python
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
juventus = fifa.loc[fifa.Club == "Juventus"]["Overall"]
#print(barcelona)
#print(madrid)
#print(juventus)
```

```python
plt.style.use("default")

labels = ["FC Barcelona","Real Madrid","Juventus"]
plt.title("Football clubs comparison")
plt.ylabel("FIFA Overall Ratings")

boxes = plt.boxplot([barcelona,madrid,juventus],labels=labels,patch_artist =True)#,color = "#abf2f5")
#patch_artist = True so that we can change face color
#here using color gives an error

for box in boxes["boxes"]:
    #for edge color
    box.set(color="#abf2f5", linewidth = 2)   
    #for face color
    box.set(facecolor="#e0e0e0")


plt.show()
```

### Scatter Plot
#### Good to use when you have 4 dimensions to your data

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii
#here we have 4 dimension x,y,bubble size and color
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```
